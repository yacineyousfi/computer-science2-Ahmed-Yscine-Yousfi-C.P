```python
# Simple Banking System with Multiple Users and Inline Comments

# Dictionary to store users and their details: username -> [password, balance]
users = {
    "user1": ["pass123", 1000.0],
    "user2": ["mypassword", 500.0],
    "admin": ["admin123", 9999.99]
}

# Allow up to 3 login attempts
attempts = 3
logged_in_user = None  # To keep track of the currently logged in user

# Login loop
while attempts > 0:
    entered_user = input("Enter username: ")  # Ask for username
    entered_pass = input("Enter password: ")  # Ask for password

    # Check if username exists and password matches
    if entered_user in users and users[entered_user][0] == entered_pass:
        print("\nLogin successful!\n")
        logged_in_user = entered_user  # Save the logged in user
        break
    else:
        attempts -= 1  # Decrease attempt count
        print(f"Incorrect credentials. Attempts left: {attempts}")
        if attempts == 0:
            print("Too many failed attempts. Exiting system.")
            exit()  # Exit the program if max attempts are reached

# Main menu loop â€” runs until user chooses to exit
while True:
    print("\n--- Main Menu ---")
    print("1. Check Balance")
    print("2. Deposit Money")
    print("3. Withdraw Money")
    print("4. Exit")

    choice = input("Choose an option (1-4): ")  # Get user's choice

    if choice == '1':
        # Show current balance
        print(f"Your current balance is: ${users[logged_in_user][1]:.2f}")

    elif choice == '2':
        # Handle deposit
        try:
            amount = float(input("Enter amount to deposit: "))  # Convert input to float
            if amount > 0:
                users[logged_in_user][1] += amount  # Add amount to balance
                print(f"${amount:.2f} deposited successfully.")
            else:
                print("Invalid amount. Please enter a positive number.")
        except ValueError:
            # Handle non-numeric input
            print("Invalid input. Please enter a number.")

    elif choice == '3':
        # Handle withdrawal
        try:
            amount = float(input("Enter amount to withdraw: "))  # Convert input to float
            if amount <= 0:
                print("Invalid amount. Please enter a positive number.")
            elif amount > users[logged_in_user][1]:
                print("Insufficient funds.")  # Check for enough balance
            else:
                users[logged_in_user][1] -= amount  # Deduct from balance
                print(f"${amount:.2f} withdrawn successfully.")
        except ValueError:
            # Handle non-numeric input
            print("Invalid input. Please enter a number.")

    elif choice == '4':
        # Exit the system
        print("Thank you for using our banking system. Goodbye!")
        break  # Break the loop and end the program

    else:
        # Handle invalid menu choices
        print("Invalid choice. Please select from 1 to 4.")
```
