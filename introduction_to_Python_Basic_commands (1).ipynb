```python
# Task 1: Assigning Values to Variables

integer_var = 10  # Storing an integer value
float_var = 3.14  # Storing a floating-point value
string_var = "Hello"  # Storing a string value

print(integer_var, float_var, string_var)  # Printing all three variables on one line, separated by commas

# Task 2: Data Types

age = 18  # Assigning an integer value to age
print(type(age))  # Printing the data type of age (should be <class 'int'>)

age = 18.5  # Changing age to a float
print(type(age))  # Printing the data type again (should be <class 'float'>)

is_student = True  # Assigning a boolean value to is_student
print(type(is_student))  # Printing the data type of is_student (should be <class 'bool'>)

# Task 3: Basic Input and Output

name = input("What is your name? ")  # Asking user for their name
print("Hello,", name + "!")  # Printing greeting with user's name

user_age = int(input("Enter your age: "))  # Asking user for their age and converting it to an integer
if user_age >= 18:
    print("You are eligible to vote.")  # Eligible to vote if age is 18 or more
else:
    print("You are not eligible to vote.")  # Not eligible if age is less than 18

# Task 4: Converting Data Types

number_str = input("Enter a number: ")  # Asking user to enter a number as a string
number_int = int(number_str)  # Converting string to integer
print("Square of the number:", number_int ** 2)  # Printing the square of the number

float_str = "4.56"  # A string representing a float
float_num = float(float_str)  # Converting string to float
print("Multiplied by 2:", float_num * 2)  # Printing result after multiplying by 2

# Task 5: Arithmetic Operators

print("Addition:", 45 + 67)  # Adding 45 and 67
print("Subtraction:", 156 - 89)  # Subtracting 89 from 156
print("Multiplication:", 12 * 7)  # Multiplying 12 by 7
print("Division:", 100 / 4)  # Dividing 100 by 4
print("Remainder:", 10 % 3)  # Finding the remainder of 10 divided by 3
print("Power:", 3 ** 4)  # 3 raised to the power of 4

# Task 6: Comparison Operators

a = 10
b = 20

print("a == b:", a == b)  # Checks if a is equal to b
print("a != b:", a != b)  # Checks if a is not equal to b
print("a < b:", a < b)  # Checks if a is less than b
print("a > b:", a > b)  # Checks if a is greater than b
print("a <= b:", a <= b)  # Checks if a is less than or equal to b
print("a >= b:", a >= b)  # Checks if a is greater than or equal to b

# Comparing two user-input numbers
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

if num1 >= num2:
    print("The first number is greater than or equal to the second.")
else:
    print("The first number is less than the second.")

# Task 7: Logical Operators

num = int(input("Enter a number: "))  # Asking for a number
if num > 10 and num < 20:
    print("The number is between 10 and 20.")  # Number is in the range (10, 20)
else:
    print("The number is NOT between 10 and 20.")

num2 = int(input("Enter another number: "))  # Asking for another number
if num2 < 5 or num2 > 50:
    print("The number is either less than 5 or greater than 50.")  # Logical OR condition
else:
    print("The number is between 5 and 50.")

# Task 8: Conditional Statements

age_check = int(input("Enter your age: "))  # Asking for age
if age_check >= 18:
    print("You are an adult.")  # Adult if 18 or older
else:
    print("You are a minor.")  # Minor otherwise

number = int(input("Enter a number: "))  # Asking for a number
if number % 3 == 0 and number % 5 == 0:
    print("FizzBuzz")  # Divisible by both 3 and 5
elif number % 3 == 0:
    print("Fizz")  # Divisible by 3 only
elif number % 5 == 0:
    print("Buzz")  # Divisible by 5 only
else:
    print(number)  # Not divisible by 3 or 5

# Task 9: Loops

# For loop from 1 to 10
for i in range(1, 11):
    print(i)

# While loop for even numbers from 2 to 20
even = 2
while even <= 20:
    print(even)
    even += 2  # Incrementing by 2 to get the next even number

# Print all numbers divisible by 4 up to n
n = int(input("Enter a number: "))
for i in range(1, n + 1):
    if i % 4 == 0:
        print(i)

# Bonus Task: Nested Loops

# Printing a triangle pattern with stars
for i in range(1, 6):  # Outer loop controls number of rows
    for j in range(i):  # Inner loop prints stars in each row
        print("*", end="")  # Print stars on the same line
    print()  # Move to the next line after each row
```
